## Identify Missing Values

### Subtask:
Identify and count the number of missing values in each column of the DataFrame.


**Reasoning**:
To identify and count missing values, I will use the `.isna()` method to find nulls and then `.sum()` to count them per column. The results will be stored in `missing_values_count` and displayed.



```python
missing_values_count = df.isna().sum()
print("Missing values per column:")
print(missing_values_count)
```

**Reasoning**:
The previous code failed because the DataFrame `df` was not defined. I need to load a dataset into a DataFrame named `df` first. I will load the `california_housing_train.csv` file into `df` and then proceed to identify and count missing values as per the subtask instructions.



```python
import pandas as pd

df = pd.read_csv('/content/sample_data/california_housing_train.csv')

missing_values_count = df.isna().sum()
print("Missing values per column:")
print(missing_values_count)
```

## Calculate Missing Percentage

### Subtask:
Calculate the percentage of missing values for each column to understand the extent of missingness.


**Reasoning**:
To calculate the percentage of missing values, I will divide the `missing_values_count` by the total number of rows in the DataFrame and multiply by 100, then store and print the result.



```python
total_rows = len(df)
missing_percentage = (missing_values_count / total_rows) * 100
print("Percentage of missing values per column:")
print(missing_percentage)
```

## Visualize Missing Data

### Subtask:
Generate a visualization (e.g., a bar chart or heatmap) to clearly show the distribution and amount of missing data across columns.


**Reasoning**:
To visualize the missing data, I will use `matplotlib` and `seaborn` to create a bar plot of the `missing_percentage` Series, setting appropriate labels, title, and y-axis limits as requested in the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
missing_percentage.plot(kind='bar')
plt.title('Percentage of Missing Values Per Column')
plt.xlabel('Columns')
plt.ylabel('Missing Percentage (%)')
plt.ylim(0, 5) # Set y-axis limit to clearly show that all values are zero
plt.show()
```
